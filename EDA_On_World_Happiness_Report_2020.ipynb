### DATA VISUALIZATION:OBJECTIVES
1. Discover the relationship between ladder_score(happiness) and GDP per capita based on region.
2. Discover the most and least happy countries.
3. Discover the most and least corrupt countries
4. Investigate the relationship between happiness and social support based on region.

```python
#set the style for the chart
sns.set_style('darkgrid')
#set figure size
plt.figure(figsize=(10,5))
#set title
plt.title('Relationship between Happiness and GDP')
sns.scatterplot(x='ladder_score', y='explained_by_log_gdp_per_capita', hue='regional_indicator', data=happiness_data, s=150)
plt.legend(bbox_to_anchor=(1.02,1),loc='upper left', borderaxespad= 0)
plt.xlabel('Ladder Score', fontweight='bold')
plt.ylabel('GDP', fontweight='bold')
```

```python
#set the style for the chart
sns.set_style('darkgrid')
#set figure size
plt.figure(figsize=(10,5))
#set title
plt.title('Relationship between Happiness and Social Support')
sns.scatterplot(x='social_support', y='ladder_score', hue='regional_indicator', data=happiness_data, s=150)
plt.legend(bbox_to_anchor=(1.02,1),loc='upper left', borderaxespad= 0)
plt.xlabel('Social Support', fontweight='bold')
plt.ylabel('Ladder Score', fontweight='bold')
```

```python
#set the style for the chart
sns.set_style('darkgrid')
#set figure size
plt.figure(figsize=(10,5))
#set title
plt.title('Relationship between Happiness and Freedom to make Life Choices')
sns.regplot(x='freedom_to_make_life_choices', y='ladder_score',data=happiness_data)
plt.xlabel('Freeom', fontweight='bold')
plt.ylabel('Ladder Score', fontweight='bold');
```

```python
#set the style for the chart
sns.set_style('darkgrid')
#set figure size
plt.figure(figsize=(10,5))
#set title
plt.title('Relationship between Happiness and Healthy Life Expectancy')
sns.regplot(x='healthy_life_expectancy', y='ladder_score',data=happiness_data)
plt.xlabel('Healthy Life Expectancy', fontweight='bold')
plt.ylabel('Ladder Score', fontweight='bold');
```

```python
#top 20 most corrupt countries

country = happiness_data.sort_values(by='perceptions_of_corruption').tail(20)
plt.figure(figsize=(15,7))
my_colors=['green','blue','purple','cyan','red','teal','brown','salmon','yellow','orange']
plt.bar(country['country_name'],country['perceptions_of_corruption'], color=my_colors)
plt.title('Top 20 Most Corrupt Countries ',fontsize=14)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Corruption Index', fontsize=14)
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

```python
#least 20 most corrupt countries

country = happiness_data.sort_values(by='perceptions_of_corruption').head(20)
plt.figure(figsize=(15,7))
my_colors=['green','blue','purple','cyan','red','teal','brown','salmon','yellow','orange']
plt.bar(country['country_name'],country['perceptions_of_corruption'], color=my_colors)
plt.title('Least 20 Most Corrupt Countries ',fontsize=14)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Corruption Index', fontsize=14)
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

```python
h_country = happiness_data.sort_values(by='ladder_score').tail(20)
u_country = happiness_data.sort_values(by='ladder_score').head(20)
plt.figure(figsize=(15,7))
my_colors=['green','blue','purple','cyan','red','teal','brown','salmon','yellow','orange']

plt.bar(u_country['country_name'],u_country['ladder_score'], color=my_colors)
plt.title('Least 20 Happy Countries ',fontsize=14)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Ladder_score', fontsize=14)
plt.grid(True)
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(15,7))
plt.bar(h_country['country_name'],h_country['ladder_score'], color=my_colors)
plt.title('Top 20 Most Happy Countries ',fontsize=14)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Ladder_score', fontsize=14)
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

```python
h_gdp = happiness_data.sort_values(by='logged_gdp_per_capita').tail(10)
l_gdp = happiness_data.sort_values(by='logged_gdp_per_capita').head(10)
plt.figure(figsize=(15,7))
my_colors=['green','blue','purple','cyan','red','teal','brown','salmon','yellow','orange']

plt.bar(h_gdp ['country_name'],h_gdp ['logged_gdp_per_capita'], color=my_colors)
plt.title('Top 10 countries with the Highest GDP ',fontsize=14)
plt.xlabel('Country', fontsize=14)
plt.ylabel('logged_gdp_per_capita', fontsize=14)
plt.grid(True)
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(15,7))
plt.bar(l_gdp['country_name'],l_gdp['logged_gdp_per_capita'], color=my_colors)
plt.title('10 Countries with the Lowest GDP ',fontsize=14)
plt.xlabel('Country', fontsize=14)
plt.ylabel('logged_gdp_per_capitare', fontsize=14)
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

### DATA VISUALIZATION: KEY TAKEAWAY
1. The datset covers 10 unique regions and 153 countries.
2. There is a positive relationship(correlation) between happiness and GDP.
    - Most of the countries in the Western Europe have very high ladder score and GDP
3. There is a positive relationship between happiness and social support.
4. Healthy life expantancy has a positive influence on happiness level.
5. Freedom to make life choices also have a positive imapct on happiness level.
6. The Western Europe have the highest GDP.
7. Most of the countries fall under the Sub Saharan Africa region.
8. The most corrupt region is the Sub Saharan Africa region with a corruption percentage of 26.8%, while the least corrupt is the North America and ANZ with a corruption percentage of 1.5%
9. The least corrupt country is Singapore, and the most is Bulgaria.
10. The most happy country is Finland, while the least happy country is Afghanistan.
11. The country with the highest GDP is Luxembourg, while the country with the lowest GDP is Burundi.

### CONCLUSIONS
1. From the findings above, it was discovered that the following have positive impact on people's happiness;
- Social support
- Freedom to make life choices
- Healthy life expectancy.

2. It was also discovered that corruption has a negative impact on people's happiness. When the level of corruption is high, the happiness of the people decreases.


3. The happier the people, the higher the GDP.
